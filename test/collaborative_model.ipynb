# 1. Generating Data
Before we create our model and make some recommendations, we need to have some user and stock data. There are no datasets that contain data with user investment accounts and their transactions so we will have to make our own.

```python
import names
import random
from pprint import pprint
from iexfinance.stocks import Stock
```

## User Accounts
There is a lot of information brokerage platforms store on the user but since not all the information will be used we will only be generating the following:
* username
* firstname
* lastname
* risk profile
* cash balance

```python
# generate random names for users
str_names = [names.get_full_name() for i in range(10000)]

# derive username, firstname and lastname from generated names
users = []
for name in str_names:
    split_name = name.split(' ')
    
    username = split_name[0].lower() + split_name[1].lower()
    firstname = split_name[0]
    lastname = split_name[1]

    user = [username, firstname, lastname]
    users.append(user)

# randomly assign a risk status to each user
# risk_status = ('Low', 'High')
risk_status = (2, 1)
for user in users:
    user.append(risk_status[random.randint(0,1)])

# randomly allocate cash amount to each user within 5000 and 100000
for user in users:
    user.append(round(random.uniform(5000,100000),2))
pprint(users)
```

### Low Profile
A low risk profile will have one stock allocated in each sector which will be selected randomly. The current cash balance will account for 50% of the porfolio. The remaining 50% will be calculated and used to determine the number of shares to hold for each stock. Divide the 50% by the number of sectors to derive the dollar amount to allocate each sector. For each stock chosen, the current stock price will be looked up. Buy the number of shares that total to less than or equal to the derived dollar amount.

### High Risk Profile
A high risk profile will have a portfolio consisting of 20% cash and 80% stocks. Of the 80% there will be two stocks with 40% allocation from two random sectors.

```python
stocks = (
    ('TD', 'Financial'),
    ('JPM', 'Financial'),
    ('V', 'Financial'),
    ('BRK.B', 'Financial'),
    ('AAPL', 'Technology'),
    ('MSFT', 'Technology'),
    ('AMZN', 'Technology'),
    ('FB', 'Technology'),
    ('ENB', 'Energy'),
    ('TRP', 'Energy'),
    ('XOM', 'Energy'),
    ('CVX', 'Energy'),
    ('BA', 'Transportation'),
    ('UNP', 'Transportation'),
    ('UTX', 'Transportation'),
    ('TSLA', 'Transportation'),
    ('PG', 'Consumer'),
    ('KO', 'Consumer'),
    ('MCD', 'Consumer'),
    ('WMT', 'Consumer')
    )

sectors = ('Financial', 'Technology', 'Energy', 'Transportation', 'Consumer')
```

```python
user_stocks = []
for id, user in enumerate(users):
    risk_profile = user[3]
    cash_bal = user[4]
    holdings = [id, risk_profile, cash_bal]
    for i in range(len(stocks)):
        holdings.append(0)
    
    
    if risk_profile == 2: # 2 = low risk
        # pick a stock from each sector
        # [financial_stock, technology_stock, energy_stock, transportation_stock, consumer_stock]
        picks = [random.randint(0,3), random.randint(4,7), random.randint(8,11), random.randint(12,15), random.randint(16,19)]
        
        # calculate dollar amount per sector
        per_sector_alloc = cash_bal/.5 - cash_bal
        
        # get the current market price for each stock and get number of shares
        for pick in picks:
            symbol = stocks[pick][0]
            price = Stock(symbol).get_price()
            num_shares = int(per_sector_alloc/price)
            holdings[pick+3] = num_shares
            
    elif risk_profile == 1: # 1 = high risk
        # select 2 sectors at random
        rand_sectors = set()
        while len(rand_sectors) < 2:
            rand_sectors.add(random.choice(sectors))
        
        # select 2 random stocks in the 2 random sectors
        picks = set()
        while len(picks) < 2:
            stock = random.choice(stocks)
            if stock[1] in rand_sectors:
                picks.add(stocks.index(stock))
                rand_sectors.discard(stock[1])
        picks = tuple(picks)
        
        # calculate dollar amount per sector
        per_sector_alloc = (cash_bal/.2 - cash_bal)/2
        
        # get the current market price for each stock and get number of shares
        for pick in picks:
            symbol = stocks[pick][0]
            price = Stock(symbol).get_price()
            num_shares = int(per_sector_alloc/price)
            holdings[pick+3] = num_shares
    
    user_stocks.append(holdings)

for user in user_stocks:
    print(user)
```
